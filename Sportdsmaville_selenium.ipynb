```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from datetime import datetime, timedelta
import time

def date():
    # Mapping anglais → français
    mois_fr = {
        "January": "janvier",
        "February": "février",
        "March": "mars",
        "April": "avril",
        "May": "mai",
        "June": "juin",
        "July": "juillet",
        "August": "août",
        "September": "septembre",
        "October": "octobre",
        "November": "novembre",
        "December": "décembre"
    }

    # Date de départ
    date_depart = datetime.now()
    
    # Ajouter 45 jours
    date_updatee = date_depart + timedelta(days=45)
    
    # Récupérer le mois en anglais
    mois_anglais = date_updatee.strftime('%B')
    jour = date_updatee.strftime('%d')
    
    # Convertir en français et construire la chaîne
    date_str = f"{mois_fr[mois_anglais]} {jour},"
    
    return date_str

def main():
    # Configuration Chrome
    chrome_options = Options()
    # chrome_options.add_argument('--headless')  # Décommentez pour mode headless
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--window-size=1920,1080')
    
    # Initialiser le driver
    driver = webdriver.Chrome(options=chrome_options)
    wait = WebDriverWait(driver, 10)
    
    try:
        # Naviguer vers la page
        driver.get("https://sport-dans-la-ville.doinsport.club/select-booking?guid=%221ce2c55d-6010-4f45-9b6f-1aafc04382fa%22&from=sport&activitySelectedId=%22cc4da804-1ef4-4f57-9fa4-4c203cdc06c8%22&categoryId=%22910503af-d67a-4f2b-a0df-838e0b4fb8ac%22")
        
        # Cliquer sur le logo calendrier
        calendar_icon = driver.find_element(By.XPATH, '//*[@id="main-content"]/app-club-booking-list/app-club-booking-header/div/app-date-select/div/div[2]/ion-button')
        calendar_icon.click()
        
        # Avancer de 2 mois
        for _ in range(3):
            next_month_btn = driver.find_element(By.XPATH, '//*[@id="main-content"]/app-club-booking-list/app-club-booking-header/div/app-date-select/ion-calendar/div/ion-button[3]')
            next_month_btn.click()
            time.sleep(0.5)  # Attendre que le calendrier se mette à jour
        
        # Cliquer sur la date
        date_to_click = date()
        try:
            # Essayer plusieurs sélecteurs possibles pour la date
            date_selectors = [
                f"//button[contains(@aria-label, '{date_to_click}')]",
                f"//button[contains(text(), '{date_to_click}')]",
                f"//*[contains(text(), '{date_to_click}')]"
            ]
            
            date_clicked = False
            for selector in date_selectors:
                try:
                    date_element = wait.until(EC.element_to_be_clickable((By.XPATH, selector)))
                    date_element.click()
                    date_clicked = True
                    break
                except:
                    continue
                    
            if not date_clicked:
                print(f"Impossible de trouver la date : {date_to_click}")
                return
                
        except Exception as e:
            print(f"Erreur lors du clic sur la date : {e}")
            return
        
        # Naviguer jusqu'à la tranche horaire 16:00 - 20:00
        found_timeslot = False
        
        for direction in ['right', 'left']:
            if found_timeslot:
                break
                
            for _ in range(5):
                try:
                    # Cliquer sur la flèche de navigation
                    arrow_btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f"button.btn-arrow-{direction}")))
                    arrow_btn.click()
                    time.sleep(1)  # Attendre que le contenu se charge
                    
                    # Vérifier le créneau
                    creneau_element = driver.find_element(By.CSS_SELECTOR, "div.value")
                    creneau_text = creneau_element.text.lower()
                    
                    if "16:00 - 20:00" in creneau_text:
                        # Cliquer sur 19:00
                        hour_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), '19:00')] | //*[contains(text(), '19:00')]")))
                        hour_btn.click()
                        found_timeslot = True
                        break
                        
                except Exception as e:
                    print(f"Erreur lors de la navigation : {e}")
                    continue
        
        if not found_timeslot:
            print("Impossible de trouver le créneau 16:00 - 20:00")
            return
        
        # Rechercher les terrains par ordre de préférence
        numero_terrain = [4, 5, 7, 6, 3, 2, 1]
        terrain_found = False
        
        for i in numero_terrain:
            try:
                # Trouver l'élément du terrain
                terrain_element = driver.find_element(By.XPATH, f"//*[contains(text(), 'Foot {i} Football 5vs5 - Exté')]")
                terrain_text = terrain_element.text
                print(f"Terrain {i}: {terrain_text}")
                
                if "Début19:00" in terrain_text and "60" in terrain_text:
                    # Cliquer sur le terrain avec 60 min
                    terrain_card = driver.find_element(By.XPATH, f"//app-card-playground[contains(., 'Foot {i} Football 5vs5 - Exté')]//ion-label[contains(., '60 min')]")
                    terrain_card.click()
                    terrain_found = True
                    break
                    
            except NoSuchElementException:
                print(f"Terrain {i} non trouvé ou non disponible")
                continue
        
        if not terrain_found:
            print("Aucun terrain disponible avec les critères recherchés")
            return
        
        # Processus de réservation
        try:
            # Remplir l'email
            email_input = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'input[placeholder="john.doe@example.com"]')))
            email_input.clear()
            email_input.send_keys('jolie.mountain@gmail.com')
            
            # Valider l'email
            validate_email_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Valider mon email')] | //*[contains(text(), 'Valider mon email')]")))
            validate_email_btn.click()
            
            # Remplir le mot de passe
            password_input = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'input[placeholder="******"]')))
            password_input.clear()
            password_input.send_keys('Toulouse31')
            
            # Valider le mot de passe
            validate_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Valider')] | //*[contains(text(), 'Valider')]")))
            validate_btn.click()
            
            # Cliquer sur Suivant
            next_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Suivant')] | //*[contains(text(), 'Suivant')]")))
            next_btn.click()
            
            # Payer et réserver (première occurrence)
            pay_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Payer et réserver')] | //*[contains(text(), 'Payer et réserver')]")))
            pay_btn.click()
            
            # Ajouter une carte
            add_card_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Ajouter une carte')] | //*[contains(text(), 'Ajouter une carte')]")))
            add_card_btn.click()
            
            # Sélectionner le radio button
            radio_btn = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "#ion-overlay-6 ion-radio")))
            radio_btn.click()
            
            # Cliquer sur Sélectionner
            select_btn = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Sélectionner')] | //*[contains(text(), 'Sélectionner')]")))
            select_btn.click()
            
            # Payer et réserver (deuxième occurrence)
            final_pay_btns = driver.find_elements(By.XPATH, "//button[contains(text(), 'Payer et réserver')] | //*[contains(text(), 'Payer et réserver')]")
            if len(final_pay_btns) > 1:
                final_pay_btns[1].click()
            else:
                final_pay_btns[0].click()
            
            # Attendre et pause
            time.sleep(1)
            print("Processus de réservation terminé. Vérifiez manuellement.")
            input("Appuyez sur Entrée pour fermer le navigateur...")
            
        except Exception as e:
            print(f"Erreur durant le processus de réservation : {e}")
            input("Appuyez sur Entrée pour fermer le navigateur...")
    
    except Exception as e:
        print(f"Erreur générale : {e}")
    
    finally:
        driver.quit()

if __name__ == "__main__":
    main()
```
